```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
data=load_iris()
print(data)
```

```python
feature=data.data
target=data.target
```

```python
std=StandardScaler()
stdd=std.fit_transform(feature)
```

```python
print(stdd)
```

```python
pca=PCA(n_components=2)
Pca=pca.fit_transform(stdd)
print(Pca)
```

```python
df=pd.DataFrame(data=Pca,columns=['PC1','PC2'])
df['target']=data.target
print(df)                              
```

```python
colors = ['red','green','blue']
plt.figure(figsize = (5,4))
for label, color in zip(data.target_names,colors):
        plt.scatter(
        df.loc[df['target'] == list(data.target_names).index(label),'PC1'],
        df.loc[df['target'] == list(data.target_names).index(label),'PC2'],
        label = label,
        color=color,
        alpha=0.5
        )
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='target')
plt.title('Principal Component Analysis')
plt.show()
```

```python
variance=pca.explained_variance_ratio_
print("PC1 ", variance[0])
print("PC2 ", variance[1])
print("Total Variance ", sum(variance))
```
